<strong> Case: Score de crédito de clientes.

Você foi contratado por um banco para consefguir definir o score de créditos dos clientes.

Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações dos clientes e dizer automaticamente o score de crédito dele:
* Ruim,
* Médio,
* Bom.

## Importação de bibliotecas:

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

## Importação dos dados:

```python
df = pd.read_csv('clientes.csv')
```

```python
display(df)
```

```python
display(df.info())
```

## Preparar a base de dados para a inteligência artificial:

Vamos transformar todas as colunas Dtype: Object em nº,  aplicando o codificador "LabelEncoder" que é uma classe da biblioteca "scikit-learn", utilizada para codificar rótulos categóricos em valores numéricos.

```python
codificador = LabelEncoder()
```

```python
df['profissao'] = codificador.fit_transform(df['profissao'])
df['mix_credito'] = codificador.fit_transform(df['mix_credito'])
df['comportamento_pagamento'] = codificador.fit_transform(df['comportamento_pagamento'])
```

```python
display(df.info())
```

## Separar as informações em dados de treino e dados de teste:

```python
#teste
y = df['score_credito']
```

```python
# retirar as colunas que não vamos usar para fazer a previsão do score de crédito (treino):
x = df.drop(['score_credito', 'id_cliente'], axis=1)

# axis =1 -> coluna
# axis =0 -> linha
```

```python
# essa ordem é obrigatória
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3, random_state=1)
```

## Criar um modelo de IA:

```python
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino);
```

## Escolher o melhor modelo:

```python
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```

O modelo_arvoredecisao é o melhor modelo pois tem 83% de acurácia.
